```python
from transformers import pipeline
import time
import torch

# Define models
models = {
    "facebook/bart-large-cnn": pipeline("summarization", model="facebook/bart-large-cnn"),
    "google/pegasus-xsum": pipeline("summarization", model="google/pegasus-xsum"),
    "t5-small": pipeline("summarization", model="t5-small"),
    "sshleifer/distilbart-cnn-12-6": pipeline("summarization", model="sshleifer/distilbart-cnn-12-6"),
}

# Sample text
text = """
Artificial intelligence (AI) is intelligence demonstrated by machines, as opposed to the natural intelligence displayed by animals and humans.
Leading AI textbooks define the field as the study of intelligent agents: any system that perceives its environment and takes actions
that maximize its chance of achieving its goals.
"""

# Store results
results = []

for model_name, summarizer in models.items():
    start_time = time.time()
    summary = summarizer(text, max_length=100, min_length=30, do_sample=False)[0]['summary_text']
    end_time = time.time()

    inference_time = end_time - start_time
    model_size = sum(p.numel() for p in summarizer.model.parameters()) * 4 / (1024**2)  # Convert to MB

    results.append({
        "Model": model_name,
        "Summary": summary,
        "Inference Time (s)": round(inference_time, 3),
        "Model Size (MB)": round(model_size, 2),
    })

import pandas as pd
df = pd.DataFrame(results)
print(df)
```

```python
from datasets import load_dataset
from evaluate import load
from transformers import AutoTokenizer

# Load dataset
dataset = load_dataset("cnn_dailymail", "3.0.0")
reference_summary = dataset["test"][0]["highlights"]
sample_text = dataset["test"][0]["article"]

# Load evaluation metric
rouge = load("rouge")

# Function to split text into tokenized chunks correctly
def chunk_text(text, model_name):
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    max_tokens = min(512, tokenizer.model_max_length)  # Get model-specific token limit
    tokens = tokenizer.encode(text, truncation=False)  # Tokenize full text

    # Split into chunks of max_tokens size
    chunks = [tokens[i:i + max_tokens] for i in range(0, len(tokens), max_tokens)]

    # Convert token chunks back to text
    return [tokenizer.decode(chunk, skip_special_tokens=True) for chunk in chunks]

# Process each model
for result in results:
    model_name = result["Model"]

    # Chunk text properly using model's tokenizer
    text_chunks = chunk_text(sample_text, model_name)

    # Generate summaries for each chunk
    generated_summaries = []
    for chunk in text_chunks:
        try:
            summary = models[model_name](chunk, max_length=min(130, len(chunk)//2), min_length=30, do_sample=False)[0]['summary_text']
            generated_summaries.append(summary)
        except Exception as e:
            print(f"Error in {model_name}: {e}")

    generated_summary = " ".join(generated_summaries)  # Merge all chunk summaries

    # Compute ROUGE score
    rouge_scores = rouge.compute(predictions=[generated_summary], references=[reference_summary])

    result["ROUGE-1"] = round(rouge_scores["rouge1"], 3)
    result["ROUGE-2"] = round(rouge_scores["rouge2"], 3)
    result["ROUGE-L"] = round(rouge_scores["rougeL"], 3)

df = pd.DataFrame(results)
print(df)
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Select relevant columns
criteria = df[["ROUGE-1", "ROUGE-2", "ROUGE-L", "Inference Time (s)", "Model Size (MB)"]].values

# Normalize (higher is better for ROUGE, lower is better for time/size)
scaler = MinMaxScaler()
normalized_criteria = scaler.fit_transform(criteria)

# Define weights for each criterion (adjust as needed)
weights = np.array([0.3, 0.3, 0.2, 0.1, 0.1])

# Calculate weighted normalized decision matrix
weighted_criteria = normalized_criteria * weights

# Find ideal best and worst
ideal_best = np.max(weighted_criteria, axis=0)  # Best for benefit criteria
ideal_worst = np.min(weighted_criteria, axis=0) # Worst for cost criteria

# Compute Euclidean distances to ideal best and worst
dist_best = np.linalg.norm(weighted_criteria - ideal_best, axis=1)
dist_worst = np.linalg.norm(weighted_criteria - ideal_worst, axis=1)

# Compute TOPSIS scores
topsis_scores = dist_worst / (dist_best + dist_worst)

df["TOPSIS Score"] = topsis_scores
df = df.sort_values(by="TOPSIS Score", ascending=False)  # Higher score is better

print(df)
```
